```python
italydf=pd.read_csv('italy-covid-daywise.csv')
```

```python
italydf
```

```python
italydf.date
```

```python
italydf.date=pd.DatetimeIndex(italydf.date)
```

```python
italydf.date
```

```python
italydf['Months']=pd.DatetimeIndex(italydf.date).month
italydf
```

```python
# now we can find graph of newcases vs new deaths at diffrent month
```

```python
plt.figure(figsize=(22 , 15))
sns.scatterplot(x = italydf.new_cases , y = italydf.new_deaths ,
               hue = italydf.Months , s=100 )
```

# Histrograms

```python
italydf.describe()
```

```python
plt.title("Drawing Histrogram")
plt.hist(italydf.new_cases)
```

```python
plt.figure(figsize=(20 , 8))
plt.hist(italydf.new_cases , bins=5)
```

```python
plt.title("Histrogram")
plt.figure(figsize=(25 , 6))
plt.hist(italydf.new_cases , bins=np.arange(0 ,2500 ,500));
```

```python
greaterdata=italydf.new_cases > 3000
greaterdata
```

```python
italydf[greaterdata]
```

```python
plt.title("Histrogram")
plt.figure(figsize=(25 , 6))
plt.hist(italydf.new_cases , bins=np.arange(0 ,7000 ,500));
```

```python
firstmonth=italydf[italydf.Months == 8]
secondmonth=italydf[italydf.Months == 3]
thirdmonth = italydf[italydf.Months == 4]
firstmonth
```

```python
plt.hist(firstmonth.new_cases , alpha=0.6)
plt.hist(secondmonth.new_cases , alpha=0.4)
plt.hist(thirdmonth.new_cases , alpha= 0.4)
plt.legend(['8th Month' , '3rd month' ,'4th month'])
```

```python
plt.hist([firstmonth.new_cases ,secondmonth.new_cases ,thirdmonth.new_cases] , stacked=True)
plt.legend(["8th Month" ,"3rd Month" , "4th month"])
```

# Bar chart
Bar chart is similar to line chart but bar chat give more detailed information about the as compared to line chart..Bar chat are more easy to understand as compared to line chart

```python
plt.bar(italydf.Months , italydf.new_cases )
plt.bar(italydf.Months , italydf.new_deaths , bottom=italydf.new_cases)
```
