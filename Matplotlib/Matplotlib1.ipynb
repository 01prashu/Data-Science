```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
years = [2010, 2011, 2012, 2013, 2014 , 2015]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]

plt.plot(yield_apples);
```

```python
plt.plot(years , yield_apples);
```

```python
yield_orange=[0.995, 0.1, 0.19, 0.26, 0.9, 0.31]
plt.plot(years, yield_orange);
plt.plot(years , yield_apples);
plt.xlabel('Year')
plt.ylabel('yield tons per hectare')
plt.title("Crop Yield in Kanto ")
plt.legend(['Apple' , 'Orange']);
```

 # line makers
  ##### we can show marker to data point marker help to understand where they actually present in chart line

```python
plt.plot(years , yield_orange , marker='o');
plt.plot(years , yield_apples , marker='x');
plt.legend(['Apple' , 'Orange']);
plt.title("Yield of City")
plt.xlabel("Year")
plt.ylabel("Crop Yield")
```
